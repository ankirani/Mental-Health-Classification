#3. Exploratory Data Analysis

###3.1 Presentase Tingkat Stres Mahasiswi Akhir

```python
# #giving blue color
# label_order = ['Normal', 'Mild', 'Moderate', 'Severe', 'Extremely Severe']

# #create a color palette using seaborn
# colors = sns.color_palette("Blues_r", n_colors=len(label_order))
```

```python
df['Stress'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Presentase Tingkat Stres Mahasiswi')
plt.show()
```

Presentase Tingkat Stres terbesar adalah Normal dengan 36.9% dan terkecil adalah Extremely Severe sebanyak 2.4%

```python
stress_counts = df.groupby(['Program Studi', 'Stress']).size().unstack().reset_index()
stress_counts.plot(kind='bar', x='Program Studi', stacked=True)
plt.title('Jumlah Mahasiswa Berdasarkan Program Studi dan Tingkat Stres')
plt.xlabel('Program Studi')
plt.ylabel('Jumlah Mahasiswa')
plt.legend(bbox_to_anchor=(1.05, 1), loc = 'upper left', title='Tingkat Stres')
plt.show()
```

Kategori tingkat stres yang paling banyak ditemukan di setiap Program Studi adalah Normal dan Severe. Sebagian besar Program Studi menunjukkan rata-rata tingkat stres pada kategori Severe (Parah). Kategori Extremely Severe (Sangat Parah) ditemukan di Program Studi Farmasi, Pendidikan Agama Islam, Tadris Bahasa Inggris, dan Teknik Informatika.

###3.2 Presentase Tingkat Kecemasan Mahasiswi Akhir

```python
df['Anxiety'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Presentase Tingkat Kecemasan Mahasiswi')
plt.show()
```

Presentase Kategoti Tingkat Kecemasan Paling Tinggi adalah Extremely Severe sebanyak 30.9%. Hal ini menunjukkan banyaknya mahasiswi yang mengalami gangguan kesehatan mental yaitu Kecemasan.

```python
anxiety_counts = df.groupby(['Program Studi', 'Anxiety']).size().unstack().reset_index()
anxiety_counts.plot(kind='bar', x='Program Studi', stacked=True)
plt.title('Jumlah Mahasiswa Berdasarkan Program Studi dan Tingkat Kecemasan')
plt.xlabel('Program Studi')
plt.ylabel('Jumlah Mahasiswa')
plt.legend(bbox_to_anchor=(1.05, 1), loc = 'upper left', title='Tingkat Kecemasan')
plt.show()
```

Kategori tingkat kecemasan yang paling banyak ditemukan di setiap Program Studi adalah Extremely Severe dan Normal. Sebagian besar Program Studi menunjukkan rata-rata tingkat stres pada kategori Extremely Severe (Paling Parah). Kategori Extremely Severe (Sangat Parah) ditemukan di Program Studi Teknik Informatika, Tadris Bahasa Inggris, dan Pendidikan Agama Islam.

###3.6 Class Distribution for Each Factor

```python
# prompt: class distribution for multiple variable : role of supervisor, literature review, analysis method, support system, self management, and  environment to be a bar diagram, just one

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your dataframe 'df' is already loaded as in your previous code

variables = ['Role of Supervisor', 'Literature Review', 'Analysis Method',
             'Support System', 'Self Management', 'Environment']

# Create the bar plot
plt.figure(figsize=(12, 6))  # Adjust figure size for better visualization

for i, var in enumerate(variables):
    plt.subplot(2, 3, i + 1)  # Create subplots for each variable
    df[var].value_counts().plot(kind='bar')
    plt.title(f'Distribusi {var}')
    plt.xlabel(var)
    plt.ylabel('Jumlah Mahasiswa')
    plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for readability
    plt.tight_layout() # Adjust layout to prevent overlapping

plt.show()
```

```python
# Assuming your dataframe 'df' is already loaded as in your previous code

variables = ['Role of Supervisor', 'Literature Review', 'Analysis Method',
             'Support System', 'Self Management', 'Environment']

# Filter for 'Bad' Literature Review
bad_literature_review_df = df[df['Literature Review'] == 'Bad']

# Class Distribution for Program Studi with 'Bad' Literature Review
plt.figure(figsize=(10, 6))
bad_literature_review_df['Program Studi'].value_counts().plot(kind='bar')
plt.title('Distribusi Program Studi (Literature Review = Bad)')
plt.xlabel('Program Studi')
plt.ylabel('Jumlah Mahasiswa')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
